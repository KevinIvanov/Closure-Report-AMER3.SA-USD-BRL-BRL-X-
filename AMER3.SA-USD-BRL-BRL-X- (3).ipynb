```python
import pandas as pd
import datetime
import yfinance as yf
from matplotlib import pyplot as plt
import mplcyberpunk
```

```python
ticket = ["AMER3.SA", "BRL=X"]  

today = datetime.datetime.now()
one_year_ago = today - datetime.timedelta(days = 365)

data_market = yf.download(ticket, one_year_ago, today)

display(data_market)
```

```python
data_closure = data_market['Adj Close']

data_closure.columns = ['Americanas S.A. (AMER3.SA)', 'Dolar']

data_closure = data_closure.dropna()

data_closure
```

```python
data_annual = data_closure.resample("Y").last()

data_monthly = data_closure.resample("M").last()

data_annual
```

```python
plt.style.use("cyberpunk")

data_closure.plot(y = "Americanas S.A. (AMER3.SA)", use_index = True, legend = False)

plt.title("Americanas S.A. (AMER3.SA)")

mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.5)

plt.savefig('Americanas.png', dpi = 300)

plt.show()
```

```python
plt.style.use("cyberpunk")

data_closure.plot(y = "Dolar", use_index = True, legend = False)

plt.title("Dolar")

mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.5)

plt.savefig('dolar.png', dpi = 300)

plt.show()
```
